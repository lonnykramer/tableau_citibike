```python
# Import dependencies
import pandas as pd
```

```python
# Reference to CSV and reading CSV into Pandas DataFrame
csv_path = "201910_tripdata.csv"
citi_df = pd.read_csv(csv_path)
citi_df
```

```python
start_station_df = citi_df.sort_values('start station id')
start_station_df.head()
```

```python
start_count = start_station_df['start station id'].value_counts()
start_count.head()
```

```python
start_count_df = pd.DataFrame(start_count)
start_count_df.reset_index(level=0, inplace=True)
start_count_df.columns=['Start Station ID', 'Nbr of Starts']
start_count_df.head(10)
```

```python
end_count = start_station_df['end station id'].value_counts()
end_count.head()
```

```python
end_count_df = pd.DataFrame(end_count)
end_count_df.reset_index(level=0, inplace=True)
end_count_df.columns=['End Station ID', 'Nbr of Ends']
end_count_df.head(10)
```
